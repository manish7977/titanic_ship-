```python
def fillage(x):
  a=x[0]
  p=x[1]
  if pd.isnull(a):
    if p==2:
      return 29
    elif p==3:
      return 24
    else:
      return 38
  else:
    return a
```

```python
df["Age"]=df[["Age","Pclass"]].apply(fillage,axis=1)
```

```python
df.isna().sum()
```

```python
df.drop("Cabin",axis=1,inplace=True)
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df.drop(["PassengerId","Name","Ticket"],axis=1,inplace=True)
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder()
df[["Sex","Embarked"]]=oe.fit_transform(df[["Sex","Embarked"]])
```

```python
df.info()
```

```python
features=df.iloc[:,1:]
target=df.iloc[:,0]
```

```python
features.columns
```

```python
f=df.drop("Survived",axis=1)
t=df["Survived"]
```

```python
f.columns
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(features,target,random_state=1,test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(xtrain,ytrain)
ypred=lr.predict(xtest)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score,roc_auc_score

ac=accuracy_score(ytest,ypred)
cm=confusion_matrix(ytest,ypred)
cr=classification_report(ytest,ypred)
roc_auc=roc_auc_score(ytest,ypred)

print(f"AC : {ac}\nR_A : {roc_auc}\nCM :\n {cm}\nCR : {cr}")
```
